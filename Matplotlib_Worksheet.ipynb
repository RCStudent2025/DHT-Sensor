## Exercise 10 â€” Fill Between (Confidence Band)
Simulate a noisy signal and a 1-Ïƒ confidence band.

**Goals:**
- Create `x` from 0 to 10 (200 points)
- Let `y = sin(x) + noise` with noise `~ Normal(0, 0.2)`
- Let `sigma = 0.3`
- Plot the line and fill the area between `y - sigma` and `y + sigma` using `plt.fill_between`


````python
# TODO: Plot a line with a confidence band using fill_between
x = np.linspace(0, 10, 200)
noise = rng.normal(0, 0.2, size=x.size)
y = np.sin(x) + noise
sigma = 0.3

# Your code here
plt.plot(x, y, label='signal')
plt.fill_between(x, y - sigma, y + sigma, alpha=0.3, label='Â±1Ïƒ')
plt.legend()
plt.show()
````

````python
def comp_country(c1, c2):
    if c1[1] == c2[1]:
        return c1[2] < c2[2]
    else:
        return c1[1] < c2[1]

def comp_country(c1, c2):
    return (c1[1], c1[2]) < (c2[1], c2[2])

def qs(country_lst):
  if len(country_lst) <= 1:
    return country_lst
  pivot = country_lst[0]
  left, right = [], []
  for item in country_lst[1:]:
    if comp_country(item, pivot):
      left.append(item)
    else:
      right.append(item)
  sort_left = qs(left)
  sort_right = qs(right)
  return sort_left + [pivot] + sort_right

with open("youth_cleaned.txt", "r") as f:
  raw_data = f.read()
  rows = raw_data.split("\n")

  countries = []
  for row in rows[1:50]:
    country, code, year = row.split(",")
    if year == "":
      pass
    else:
      countries.append((country,code, year))
    sorted_countries = qs(countries)
  print(sorted_countries)
  output = ""
  for c in countries:
    output += f"{c[0]},{c[1]},{c[2]}\n"
  with open("youth_output.txt","w") as f:
    f.write(output)

````
